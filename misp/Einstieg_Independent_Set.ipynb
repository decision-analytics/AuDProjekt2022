**Gegeben**:

**Gesucht**:
- eine maximal große stabile Menge (*independent set*) Teilmenge der Knoten in dem Graph
- eine Menge von Knoten ist stabil (*independent*), wenn zwischen den Knoten der Menge keine dirkekten Kanten existieren

## Ein Beispielgraph 

<img src="./img/mis.png" width="17%" align ="right" >

- um das Problem zu Testen, ist es in der Regel hilfreich, einen Beispielgraph zu betrachten
- wir nehmen den, den wir schon kennen:
- und wir betrachten verschiedene Möglichkeiten, einen Graphen "zu befüllen"




````python
g = nx.Graph()

# Knoten von 0 bis 6 hinzufügen
for id in range(7):        
    g.add_node(id)


# Kanten für die Nachbarn Knoten 0 hinzufügen
g.add_edge(0, 1)
g.add_edge(0, 5)
g.add_edge(0, 6)

# Weitere Kanten (in kompakterer Schreibweise)
g.add_edges_from([(1, 6), (1, 2), (2, 6), (2, 3), (3, 4), (3, 6),(4,5), (4,6), (5,6)])

# Ausgeben der Knoten und Kanten, um sie anzugucken
print('Knoten:', g.nodes())
print('Kanten:', g.edges())
````

## Graph zeichnen
- oft ist es hilfreich, einen Graphen auch zu Zeichnen
- NetworkX bietet eine Reihe von Funktionen zum Zeichnen von Graphen: 
  - https://networkx.org/documentation/stable/reference/drawing.html

````python
## alle Knoten grau (bilde ein Array, in dem für jeden Knoten 'lightgrey' steht)
node_colors = ['lightgrey' for node in g.nodes()]
# dieser drawing-Ansatz gibt uns eine schöne Abbildung
nx.draw_kamada_kawai(g, with_labels=True, node_color= node_colors)
````

- nun färben wir wie im Beispiel eine Teilmenge ein

````python
node_colors[4]= 'red'
node_colors[0]= 'red'
node_colors[2]= 'red'
nx.draw_kamada_kawai(g, with_labels=True, node_color= node_colors)
````

## Einlesen von Probleminstanzen
- beim MISP sind Probleminstanzen schlicht Graphen
- eine Menge von Instanzen für Graphen finden Sie im Ordner `instances`
  - diese können Sie auch im Projekt verwenden!
- schauen Sie sich eine solche Instanz mal mit einem Text-Editor an!

Es zeigt sich:
- wenn in einer Zeile der erste Buchstabe `p` ist, so steht dort die Anzahl der Knoten
- in jeder Zeile nach dieser Zeile finden sich eine `e` und dann die Kanten als Paar zweier Knoten
  - dabei sind Knoten von 1 bis $n$ nummeriert ($n$ = Anzahl der Knoten)


````python
def read_graph_from_col_file(filename):
    file = open(filename, 'r')
    g = nx.Graph()
    
    #Nodes
    for line in file:
        if line[0]=='p': # Anzahl der Knoten wird gesucht
            anz_knoten = int(line.split()[2])
            break
    for n in range(1,anz_knoten+1): # Richtige Anzahl an Knoten wird hinzugefügt
        g.add_node(n)
    
    #Edges
    for line in file: # Alle angegebenen Kanten werden hinzugefügt
        g.add_edge(int(line.split()[1]), int(line.split()[2]))
    
    return g
````

## Einlesen von Probleminstanzen
- hier ein Beispiel:

````python
path = './instances/myciel3.col' # Beispiel für eine der Probleminstanzen
graph_instance = read_graph_from_col_file(path)
````

..auch hier können wir den Graphen visualisieren:

````python
nx.draw_kamada_kawai(graph_instance, with_labels=True, node_color= 'lightgrey')
````
